```python
import pandas as pd
import matplotlib.pyplot as plt


df = pd.read_csv('company_sales_data.csv')

total_profit = df['total_profit']


plt.plot(total_profit, linestyle='--', color='red', marker='o', markerfacecolor='red', linewidth=3, label='Total Profit')


plt.xlabel('Month Number')
plt.ylabel('Sold Units Number')
plt.legend(loc='lower right')


plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt


df = pd.read_csv('company_sales_data.csv')



df.set_index('month_number', inplace=True)


plt.figure(figsize=(12, 8))
for column in ['facecream', 'facewash', 'toothpaste', 'bathingsoap', 'shampoo', 'moisturizer']:
    plt.plot(df.index, df[column], marker='o', label=column)

plt.xlabel('Month Number')
plt.ylabel('Units Sold')
plt.title('Monthly Units Sold per Product')
plt.legend(title='Product')
plt.grid(True)

plt.show()
```

```python
df = pd.read_csv('company_sales_data.csv')

total_units_per_product = df[['facecream', 'facewash', 'toothpaste', 'bathingsoap', 'shampoo', 'moisturizer']].sum()


plt.figure(figsize=(8, 8))
plt.pie(total_units_per_product, labels=total_units_per_product.index, autopct='%1.1f%%', startangle=140)


plt.title('Total Units Sold per Product for the Year')


plt.show()
#autopct='%1.1f%%' is a way of how the % are represented 

```
